```python
class By:
    """Set of supported locator strategies."""
    ID = "id"
    XPATH = "xpath"
    LINK_TEXT = "link text"
    PARTIAL_LINK_TEXT = "partial link text"
    NAME = "name"
    TAG_NAME = "tag name"
    CLASS_NAME = "class name"
    CSS_SELECTOR = "css selector"
```

```python
def captcha_solver(img_link):
    # Strip off the prefix if present
    header, encoded = img_link.split(",", 1)
    # Decode and write to file
    with open("captcha-img.jpg", "wb") as f:
        f.write(base64.b64decode(encoded))
    # img = Image.open(f'captcha-img.png')
    img = Image.open(f'captcha-img.jpg').convert("RGB")  # or "L" for grayscale
    # Invert the colors
    inverted = ImageOps.invert(img)
    # Save the result
    inverted.save("inverted_image.jpg")
    inverted.save("inverted_image.jpg")
    # text = pytesseract.image_to_string(inverted,config='--psm 8',) 
    text2 = pytesseract.image_to_string(img) 
    # print(text,text2)
    return text2
# captcha_solver()
```
